```python
import pandas as pd
import numpy as np
```

```python
predictions = pd.read_csv('nb36.csv', index_col=0)
test = pd.read_csv('test_translated_clean.csv', index_col=0)

def find_good_predicts(valids, column):
    b1 = valids[column] > 0.96
    b2 = valids[column] < 0.1
    c = valids[b1|b2]
    print(valids[b1].shape)
    print(valids[b2].shape)
    return c

good_predictions = find_good_predicts(predictions,'toxic')
g = good_predictions.join(test)
train = pd.read_csv('train_translated_clean.csv',index_col=0)
new_train = pd.concat([train,g])

```

```python
new_train.head(n=20)
```

```python
fn = 'train_' + str(new_train.shape[0]) + '.csv'
new_train.to_csv(fn)
```
