```python
from bigfloat import *
def logit(p):
    with precision(100000):
        return log(p)- log(1 -BigFloat(p))
```

```python
from scipy.special import expit
from scipy.special import logit
```

```python
xl = lambda t: expit(logit(t) - 0.5)
```
